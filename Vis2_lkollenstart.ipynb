```python
#final data

# import bokeh functions
from bokeh.io import output_notebook, show, push_notebook
from bokeh.plotting import figure
from bokeh.models import HoverTool, ColumnDataSource
import bokeh.tile_providers

# import interactive widgets
from ipywidgets import interact, SelectionSlider

# import pandas
import pandas as pd

# numpy sqrt
from numpy import sqrt

zika_data = pd.read_csv("/Users/Leonie/Python/day4-lkollenstart/data/zika.csv")

# create a figure
zika_fig = figure(x_range=(-11627260, -5627261), y_range=(-2724688, 3275312))

# hide the axes
zika_fig.axis.visible = False

# Add a map tile
zika_fig.add_tile(bokeh.tile_providers.STAMEN_TERRAIN);

# make a list of the years for use later
dates = sorted(list(set(zika_data.Date)))


#new circles
zika_data["size"] = 5 + sqrt(zika_data.Cases)

zika_data.head()

#add circle to figure
zika = zika_fig.circle(
    x="X", y="Y", size="size", source=zika_data,
    fill_color="red", fill_alpha=0.6,
    line_color="black", line_alpha=0.6, line_width=2
)

# define tooltip information
tooltips = HoverTool(tooltips=[
    ("Country", "@Country"),
    ("Cases", "@Cases")
]) 

# add the tooltips
zika_fig.add_tools(tooltips)

# show the plot, saving a handle for us to push to
zika_handle = show(zika_fig, notebook_handle=True)

    

def update_zika(Dates):
    zika.data_source.data = ColumnDataSource(zika_data[zika_data["Date"] == Dates]).data
    zika_fig.title.text = "Zika cases on {}".format(Dates)
    push_notebook(handle=zika_handle)

    # make the slider
zika_slider = interact(
    update_zika,
    Dates=SelectionSlider(description='Dates', options=dates)
)
```
